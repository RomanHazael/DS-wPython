```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import re
```

```python
results = pd.read_csv("./survey_results_public.csv", index_col="ResponseId")
resultsSchema = pd.read_csv("./survey_results_schema.csv")
```

```python

languages = 'LanguageHaveWorkedWith'
salary = 'ConvertedCompYearly'

#.dropna(no toma en cuenta ceros), .loc(:,:), .rename (renombra), sort_values(ordena)
resultstable = results.dropna(subset=[languages, salary]).loc[:,['Country', languages, salary ]].rename(columns={languages:'Languages', salary:'Salary'}).sort_values(by = 'Salary')

#filtros
salaryfilter = (results[salary]>=10000) & (results[salary]<=1e6)
mexicanfilter = results['Country'] =='Mexico'
resultstable = resultstable[(salaryfilter) ]

# Para calcular cual es la linea que mas ';' tiene 
hlc = resultstable['Languages'].str.count(';') # HLC (Higher Language Count)
idmlc= hlc.idxmax()

alllanguages = resultstable['Languages'][idmlc].split(';')

for row in resultstable['Languages']:
    for lang in row.split(';'):
        if lang in alllanguages:
            continue
        else:
            alllanguages.append(lang)
   
#limpiar sin parentesis         
alllanguages = [re.sub(r'\(.*\)', '', lang) for lang in alllanguages]
alllanguages
```

```python
#Tabla inicial de Lenguaje-Salario
meanSalary = pd.DataFrame(data = np.zeros(len(alllanguages)), index=alllanguages, columns = ['Salary'])
meanSalary
```

```python

for lang in alllanguages:
    mask = resultstable['Languages'].str.contains(lang)
    average = resultstable.loc[mask,'Salary'].mean()
    meanSalary.loc[lang, 'Salary'] = average
meanSalary
```

```python
colors = [
    "#03071E",
    "#370617",
    "#6A040F",
    "#9D0208",
    "#D00000",
    "#DC2F02",
    "#E85D04",
    "#F48C06",
    "#FAA307",
    "#FFBA08"
]

fontdict = {
    'family': 'serif',
    'color':  'darkred',
    'weight': 'normal',
    'size': 16,
}
```

```python
meanSalary.sort_values(by='Salary', inplace=True)
meanSalary
```

```python
#plots
%matplotlib qt5
#plt.style.use('seaborn')
plt.barh(meanSalary.index, meanSalary['Salary'], color = colors)
plt.vlines(meanSalary['Salary'].mean(), 0, len(alllanguages), color= 'brown')
plt.xlabel('Salary in dollars')
plt.ylabel('Programming Languages')
plt.tight_layout()
plt.show()
```
